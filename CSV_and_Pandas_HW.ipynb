```python
import pandas as pd
acc_players_df = acc_players_df = pd.read_csv('/content/acc_players-2324F.csv')
acc_players_df.head(10)
```

```python
acc_players_df = acc_players_df.rename(columns={'Unnamed: 13': 'Pts'})
acc_players_df['Pts'] = pd.to_numeric(acc_players_df['Pts'], errors='coerce')
total_points = acc_players_df['Pts'].sum()
total_points
```

```python
acc_players_df = acc_players_df.rename(columns={'Unnamed: 6': 'MP'})
acc_players_df = acc_players_df.rename(columns={'Unnamed: 1': 'Player'})
acc_players_df['MP'] = pd.to_numeric(acc_players_df['MP'], errors='coerce')
minutes_played = acc_players_df[['Player', 'MP']].sort_values(by='MP', ascending=False).head(1)
print("Player with most minutes:")
minutes_played.head()
```

```python
acc_players_df = acc_players_df.rename(columns={'Unnamed: 7': 'TRB'})
acc_players_df['TRB'] = pd.to_numeric(acc_players_df['TRB'], errors='coerce')
total_rebounds = acc_players_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
total_rebounds.head()
```

```python
df_filtered = acc_players_df[acc_players_df['MP'] > 500]
df_filtered.head()
```

```python
df_filtered = df_filtered.rename(columns={'Unnamed: 8': 'AST'})
df_filtered['AST'] = pd.to_numeric(df_filtered['AST'], errors='coerce')
assist_total = df_filtered[['Player', 'AST']].sort_values(by='AST', ascending=False).head(1)
assist_total.head()
```

```python
assist_leads = df_filtered[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
assist_leads.head()
```

```python
df_filtered = df_filtered.rename(columns={'Unnamed: 10': 'BLK'})
df_filtered['BLK'] = pd.to_numeric(df_filtered['BLK'], errors='coerce')
top_shot_blockers = df_filtered[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
top_shot_blockers.head()
```

```python
acc_players_df = acc_players_df.rename(columns={'Unnamed: 4': 'School'})
points_by_school = acc_players_df.groupby('School')['Pts'].sum()
points_by_school.head()
```

```python
acc_players_df = acc_players_df.rename(columns={'Unnamed: 8': 'AST'})
acc_players_df['AST'] = pd.to_numeric(acc_players_df['AST'], errors='coerce')
assists_by_school = acc_players_df.groupby('School')['AST'].sum()
assists_by_school.head()
```

```python
points_by_school = acc_players_df.groupby('School')['Pts'].sum()
top_points_schools = points_by_school.sort_values(ascending=False).head(3)
top_points_schools.head()
```
